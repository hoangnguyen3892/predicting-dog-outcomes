### ----------------------DateTime----------------------
Objective:  Non metric variable: TimeofDay (morning, midday, lateday and night)
           Non metric variable: WeekDay (0, 1, 2, 3, 4, 5, 6)

```python
data_dog
```

```python
# Define function
def time_of_day(x):
    x = str(x)
    y = x.split()
    if len(y) == 2:
        time = y[1]
        hour = int(time.split(":")[0])
    if len(y) == 3:
        time = y[1]
        apm = y[2]
        if apm == "AM":
            hour = int(time.split(":")[0])
        if apm == "PM":
            hour = (int(time.split(":")[0]) + 12)
        
    if hour >= 5 and hour < 11: return "morning"
    if hour >= 11 and hour < 16: return "midday"
    if hour >= 16 and hour < 20: return "lateday"
    else: return "night"
```

```python
# Apply function
data_dog['TimeofDay'] = data_dog['DateTime'].apply(time_of_day)
```

```python
sns.countplot(data_dog.SexuponOutcome, palette='Set3')
```

```python
def getSex(x):
    x = str(x)
    if x.find('Male') >= 0: return "Male"
    if x.find('Female') >= 0: return "Female"
    #return "Unknown"
def getIntactness(x):
    x = str(x)
    if x.find('Neutered') >= 0: return 0
    if x.find('Spayed') >= 0: return 0
    if x.find('Intact') >= 0: return 1
    #return "Unknown"
```

### ----------------------AgeuponOutcome----------------------
Objective:  Metric variable: AgeinYears (age of animal)
            Non metric variable: Age (Juvenile, Adult, Old)

```python
def calcAgeInYears(x):
    x = str(x)
    if x == 'nan': return 0
    age = int(x.split()[0])
    if x.find('year') > -1: return age 
    if x.find('month')> -1: return age / 12.
    if x.find('week')> -1: return age / 52.
    if x.find('day')> -1: return age / 365.
    else: return 0
```

```python
data_dog['AgeInYears'] = data_dog['AgeuponOutcome'].apply(calcAgeInYears)
sns.distplot(data_dog.AgeInYears, bins = 20, kde=False)
```

```python
def Age(x):
    if x <2: return "Juvenile"
    if x >= 2 and x <= 6: return "Adult"
    else: return "Old"
```

```python
data_dog["Age"] = data_dog["AgeInYears"].apply(Age)
```

```python
def pure_mixed_hybrid(x):
    if x.find("Mix") >= 0:
        return "Mixed"
    elif x.find("/") >= 0:
        return "Hybrid"
    else:
        return "Pure"
```

```python
# Keep simple and dominant breed
def keepSimpleBreed(x):
    if x.find("/") >= 0:
        x.split("/")
        x = x.split("/")[0]
    if x.find("Mix") >= 0:
        x = x[:-4]
    return x 
```

```python
data_dog["SimpleBreed"] = data_dog["Breed"].apply(keepSimpleBreed)
```

## ----------------------Encode----------------------

```python
label_name = "OutcomeType"
ordinal_features = ["Age", "Size"]
nominal_features = ["HasName","TimeofDay", "WeekDay", "IsIntact", "Sex", "PureMixedHybrid", "Group", "IsMixColor"]
```

```python
# Encode dependence variable
le = LabelEncoder()
Animals[label_name] = le.fit_transform(Animals[label_name].values)
label_list = list(le.inverse_transform([0, 1, 2, 3]))     # List of outcomes
```

```python
label_list
```

```python
# Map ordinal independence variable
age_mapping = {"Juvenile": 0, "Adult": 1, "Old": 2}
Animals["Age"] = Animals["Age"].map(age_mapping)

size_mapping = {'S': 0, 'M': 1, 'L': 2, 'G':3}
Animals['Size'] = Animals['Size'].map(size_mapping)
```

```python
# One hot encoder for nominal features
Animals_encoded = pd.get_dummies(Animals, columns=nominal_features, drop_first=False)
```

```python
for i in nominal_features:
    Animals_encoded[i] = Animals[i]
```
