## Plot confusion matrix

```python
def plot_confusion_matrix(cm, classes, normalize=False,
                          title='Confusion matrix', cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    
    plt.title(title)
   
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)
    
    #plt.yticks(tick_marks, classes, rotation=45)
    #plt.xticks(tick_marks, classes)
    
    if normalize:
        cm = np.round(cm.astype('float') / cm.sum(axis=1)[:, np.newaxis], decimals=2)
        plt.imshow(cm, interpolation='nearest', cmap=cmap)
        plt.colorbar()
        print("Normalized confusion matrix")
    else:
        plt.imshow(cm, interpolation='nearest', cmap=cmap)
        plt.colorbar()
        print('Confusion matrix, without normalization')

    print(cm)
    
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()

# Compute confusion matrix
cnf_matrix = confusion_matrix(test_label_logit, predict)
np.set_printoptions(precision=2)
class_names = label_list
# Plot non-normalized confusion matrix
plt.figure()
plot_confusion_matrix(cnf_matrix, classes=class_names,
                      title='Confusion matrix, without normalization')
plt.savefig('fg/confusion_matrix_wo_normalize.png')
# Plot normalized confusion matrix
plt.figure()
plot_confusion_matrix(cnf_matrix, classes=class_names, normalize=True,
                      title='Normalized confusion matrix')
plt.savefig('fg/confusion_matrix_w_normalize.png')
plt.show()
plt.close()
```

```python
def confusion_probs(predictions, labels, ax = None, vmargin = 0.1, equalize = False, cmap = cm.jet, colors = [],ticklabels=[]):
    """
        Make a confusion-matrix-like scatter plot
    """
    N, D = predictions.shape
    bins = bincount(labels)
    assert labels.min() >= 0 and labels.max() < D, 'labels should be between 0 and {0:d}'.format(D - 1)
    assert predictions.min() >= 0 and predictions.max() <= 1, 'predictions should be normalized (to [0, 1])'
    """ Set up graphical stuff. """
    fig = None
    if ax is None:
        fig, ax = subplots(figsize = (7, 7))
        fig.tight_layout()
    if equalize:
        xticks = [x + .5 for x in range(D)]
        bin_left = range(D)
        bin_right = range(1,D+1)
    else:
        xticks = (float(D) / N) * (cumsum(bins) - bins / 2)
        bin_left = (float(D) / N) * (cumsum(bins) - bins)
        bin_right = (float(D) / N) * (cumsum(bins))
        
    ax.set_xticks(xticks)
    if len(ticklabels) == 0:
        ax.set_xticklabels(range(D))
    else:
        ax.set_xticklabels(ticklabels,rotation=90)
        
    ax.set_xlim([0, D])
    ax.set_yticks([(x + .5) * (1 + vmargin) for x in range(D)])
    if len(ticklabels) == 0:
        ax.set_yticklabels(range(D))
    else:
        ax.set_yticklabels(ticklabels)
    
    
    ax.set_ylim([0, D * (1 + vmargin) - vmargin])
    ax.set_xlabel('True classes')
    ax.set_ylabel('Prediction probabilities')
    """ Sort predictions by label. """
    indices = labels.argsort()
    X = predictions[indices, :]
    y = labels[indices]
    """ Add more vertical offset to higher labels. """
    X += linspace(0, D - 1, D) * (1 + vmargin)
    """ Make a new horizontal axis to give all samples equal width. """
    if equalize:
        z = hstack([k + linspace(0, 1, t) for k, t in enumerate(bins)])
    else:
        z = linspace(0, D, N)
    """ Plot the results with class-coding. """
    if len(colors) == 0:
        colors = []
        for q in range(D):
            colors.append(cmap(y / float(D)))
    for q in range(D):
        row = X[:, q]
        for qq in range(D):
            if qq == 3:
                ax.scatter(z[(z > bin_left[qq])&(z<bin_right[qq])], row[(z > bin_left[qq])&(z<bin_right[qq])], c = colors[qq], edgecolors = 'k',linewidth=0.1,s=5)
            else:
                ax.scatter(z[(z > bin_left[qq])&(z<bin_right[qq])], row[(z > bin_left[qq])&(z<bin_right[qq])], c = colors[qq],edgecolor='None',s=5)
    plt.tight_layout()
    return (fig, ax) if fig else ax

colors = ['#ffcee5','#3C7DC4','#f9fc37','#FFF5EE']
ticklabels = label_list
fig, ax = confusion_probs(predict_label, test_label_logit, equalize = True,colors = colors,ticklabels = ticklabels)
fig.savefig('fg/conf_mat_probs_equalized.png')
plt.show()
plt.close()
```

```python
test = pd.concat([test_label_logit, test_data_logit], axis=1)
```

```python
labels_mapping = {0: 'Adoption', 1: 'Euthanasia', 2: 'Return to Owner', 3: 'Transfer'}
train_logit['OutcomeType'] = train_logit['OutcomeType'].map(labels_mapping)
test['OutcomeType'] = test['OutcomeType'].map(labels_mapping)
```

```python
WeekDay_mapping = {0: 'Monday',1: 'Tuesday', 2: 'Wednesday', 3: 'Thursday', 4: 'Friday', 5: 'Saturday', 6: 'Sunday'}
binary_mapping = {0: 'No', 1: 'Yes'}
```

```python
#WeekDay
train_logit['WeekDay'] = train_logit['WeekDay'].map(WeekDay_mapping)
test['WeekDay'] = test['WeekDay'].map(WeekDay_mapping)

#HasName
train_logit['HasName'] = train_logit['HasName'].map(binary_mapping)
test['HasName'] = test['HasName'].map(binary_mapping)

#IsIntact
train_logit['IsIntact'].astype(int)
test['IsIntact'].astype(int)
train_logit['IsIntact'] = train_logit['IsIntact'].map(binary_mapping)
test['IsIntact'] = test['IsIntact'].map(binary_mapping)

#IsMixColor
train_logit['IsMixColor'] = train_logit['IsMixColor'].map(binary_mapping)
test['IsMixColor'] = test['IsMixColor'].map(binary_mapping)
```
